```python
# SERIES 

# Pandas Series Example
a = [1,7,3,4,5,6,7,8,9,10]

var = pd.Series(a)
print(var) 

# labeling the Series with custom index
myvar = pd.Series(a, index=["a", "b", "c", "d", "e", "f", "g", "h", "i", "j"])
print(myvar["c"]) 

# craeting a Series from a dictionary
calories = {"day 1" : 420, "day 2": 380, "day 3": 390, "day 4": 410, "day 5": 430}
calories_series = pd.Series(calories)
print(calories_series) 

# indexing a dictionary Series
calories_series = pd.Series(calories, index = ["day 1", "day 3"])
print(calories_series) 
```

```python
# DATA FRAME 

# Creating a DataFrame from a dictionary
data = {
    "calories": [420, 380, 390, 410, 430],
    "duration": [50, 40, 45, 35, 60]
}

myvar = pd.DataFrame(data, index=["day 1", "day 2", "day 3", "day 4", "day 5"])
print(myvar) 

print(myvar.loc["day 3"]) # Accessing a row by index
```
