```python
import pandas as pd


# This script reads a CSV file named "ODI data.csv", removes rows with any missing values, and prints the resulting DataFrame.

df = pd.read_csv("ODI data.csv") 

new_df = df.dropna() # Remove rows with any missing values

print(new_df.to_string()) 
```

```python
df = pd.read_csv("ODI data.csv") 

new_df = df.dropna(inplace=True) # Remove rows with any missing values and modify the original DataFrame 

print(new_df)  # This will print None since inplace=True modifies the original DataFrame and returns None 
```

```python
df = pd.read_csv("ODI data.csv") 

df.fillna(0, inplace=True)  # Replace NaN values with '0' 

print(df.to_string())  
```

```python
df = pd.read_csv("ODI data.csv")

df.fillna({'Player': 'Ali Zeb'}, inplace=True)  # Replace NaN values with specific values for each column

print(df.to_string())  
```

```python
df = pd.read_csv("ODI data.csv")

x = df['Runs'].mean()  # Calculate the mean of the 'Runs' column 

df.fillna({'Runs': x}, inplace=True)  # Replace NaN values in 'Runs' column with the mean value 

print(df.to_string())  
```

```python
df = pd.read_csv("ODI data.csv")

x = df['Inns'].median()  # Calculate the median of the 'Inns' column

df.fillna({'Inns': x}, inplace=True)  # Replace NaN values in 'Inns' column with the median value

print(df.to_string())  
```

```python
df = pd.read_csv("ODI data.csv")

x = df['Span'].mode()[0]  # Calculate the mode of the 'Span' column

df.fillna({'Span': x}, inplace=True)  # Replace NaN values in 'Span' column with the mode value

print(df.to_string())  
```

# CLEANING DATA  --  Cleaning Wrong Formate

```python
df = pd.read_csv("ODI data.csv")

df['Date'] = pd.to_datetime(df['Date'], format= 'mixed')  # Convert 'date' column to datetime format 

print(df.to_string())  
```

```python
df = pd.read_csv("ODI data.csv")

df.dropna(subset=['Date'], inplace=True)  # Remove rows where 'date' column has NaN values 

print(df.to_string())  
```

# CLEANING DATA  --  Cleaning Wrong Date

```python
df = pd.read_csv("ODI data.csv")

df.loc[31, 'Mat'] = 100  # Update the 'Mat' value at index 31 to 100

print(df.to_string()) 
```

```python
df = pd.read_csv("ODI data.csv")

for x in df.index:  # Iterate through each index in the DataFrame 
    if df.loc[x,'Inns'] == 0:
        df.loc[x, 'Inns'] = df['Inns'].mean()  # Update 'Inns' value to the mean where it was 0 
        
print(df.to_string())  
```

```python
df = pd.read_csv("ODI data.csv")

for x in df.index:  # Iterate through each index in the DataFrame
    if df.loc[x, 'Runs'] == 0:
        df.drop(x, inplace=True)  # Update 'Runs' value to the mean where it was 0 

print(df.to_string()) 
```
