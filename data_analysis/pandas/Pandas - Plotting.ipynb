```python
import pandas as pd

import matplotlib.pyplot as plt # importing matplotlib for plotting

df = pd.read_csv("ODI data.csv")

df.plot() #plotting the data from the CSV file

plt.show() #showing the plot 

```

```python
df = pd.read_csv("ODI data.csv")

df.plot(kind='bar') # plotting the data as a bar chart

plt.show() # showing the bar chart 
```

```python
df = pd.read_csv("ODI data.csv")

df.plot(kind='scatter', x='Mat', y='Runs') # plotting a scatter plot with 'Mat' on x-axis and 'Runs' on y-axis

plt.show() # showing the scatter plot 
```

```python
df = pd.read_csv("ODI data.csv")

df.plot(kind='scatter', x='Mat', y='100') # plotting a scatter plot with 'Year' on x-axis and '100' on y-axis

plt.show() # showing the scatter plot 
```

```python
df = pd.read_csv("ODI data.csv")

df["Runs"].plot(kind='hist') # plotting a histogram of the 'Runs' column 

plt.show() # showing the histogram 
```
