# FILTERING AND ORDERING

```python
import pandas as pd

df = pd.read_csv('world_population.csv')

df 
```

```python
df[df['Rank'] <= 10] # filtering the top 10 countries by population 
```

```python
Sp_country = ['Bangladesh','Pakistan', 'India'] # Countries to filter


df[df['Country'].isin(Sp_country)] # Filtering specific countries  
```

```python
df[df['Country'].str.contains('United')] # contains 'United' in the country name 
```

```python
df2 = df.set_index('Country') # set index to 'Country'

df2 
```

```python
df2.filter(items=['Capital', 'Continent', '2022 Population'], axis=1) # filtering columns 
```

```python
df2.filter(items=['Pakistan', 'United Kingdom'], axis=0) # filtering by index labels 
```

```python
df2.filter(items=['Pakistan', 'India'], axis=0).filter(
    items=['Capital', 'Continent', '2022 Population'], axis=1
    ) # filtering both rows and columns
```

```python
df2.filter(like='Pak', axis=0) # Filtering countries with 'Pak' in their name 
```

```python
df2.loc['Pakistan']  # Accessing a single row by label 
```

```python
df.iloc[92] # using iloc to access by index 
```

```python
df[df['Rank'] < 10].sort_values(by = 'Rank', ascending = True)   #.sort_values(by = 'Rank', ascending = True) 
```

# Groupby and Aggregation

```python
import pandas as pd

df = pd.read_csv('Flavors.csv')

df 
```

```python
Group_by = df.groupby("Base Flavor")  # Grouping by 'BaseFlavor' 
```

```python
Group_by.mean(numeric_only=True)  # Calculating the mean of each group 
```

```python
df.groupby("Base Flavor").mean(numeric_only=True)  # Calculating the mean of each group 
```

```python
df.groupby("Base Flavor").count()  # Counting the number of occurrences in each group 
```

```python
df.groupby("Base Flavor").min()
```

```python
df.groupby("Base Flavor").max()
```

```python
df.groupby("Base Flavor").sum(numeric_only=True)  # Summing the values in each group 
```

```python
df.groupby("Base Flavor").agg({"Flavor Rating": ['mean','max','min','sum']})  # Aggregating with multiple functions
```

```python
df.groupby(["Base Flavor", "Flavor"]).agg({"Flavor Rating": ['mean','max','min','sum']})
```

```python
df.groupby("Base Flavor").describe()  # Descriptive statistics for each group 
```

# Merge Join And Concatinate

```python
import pandas as pd

df1 = pd.read_csv('LOTR.csv')  # Corrected typo from 'raed_csv' to 'read_csv'

df1  
```

```python
df2 = pd.read_csv('LOTR 2.csv') 

df2 
```

```python
df1.merge(df2)  # Merging two DataFrames on common columns
```

```python
df1.merge(df2, how='inner', on = ['FellowshipID', 'FirstName'])  # Inner join on specific columns
```

```python
df1.merge(df2, how='outer')  # Outer join (default) 
```

```python
df1.merge(df2, how='left') 
```

```python
df1.merge(df2, how='right')  # Right join 
```

```python
df1.merge(df2, how='cross')  # Cross join (Cartesian product) 
```

```python
df1.join(df2, how='outer', on = 'FellowshipID', lsuffix='_left', rsuffix='_Right')  # Using join method for outer join 
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), how='outer', lsuffix='_left', rsuffix='_Right')  # Joining on index with suffixes 
df4 
```

```python
pd.concat([df1, df2], join='outer', axis=1)  # Concatenating DataFrames 
```

# Data Cleaning

```python
import pandas as pd

df = pd.read_excel('Customer Call List.xlsx')  # Reading Excel file

df 
```

```python
df = df.drop_duplicates() # Dropping duplicate rows 

df 
```

```python
df = df.drop(columns='Not_Useful_Column')  # Dropping a specific column 
```

```python
# df['Last_Name'] = df['Last_Name'].str.strip('/').str.strip('...') 

df['Last_Name'] = df['Last_Name'].str.strip('/.') # removing Vaues 

df 
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('|','').str.replace('-','').str.replace('/','').str.replace('Na','')


df 
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10]) # Formatting phone numbers
df 
```

```python
df[['Street Add','State', 'Zip']] = df['Address'].str.split(',', expand=True)  # Splitting 'Address' into multiple columns 

df = df.drop(columns='Address')  # Dropping the original 'Address' column
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y').replace('No', 'N').str.replace('N/a','')  # Replacing values in 'Paying Customer' column

df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y').replace('No', 'N')

df.fillna('', inplace=True)  # Filling NaN values with empty strings

 

df 
```

```python
df 
```

```python
for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == 'Y':
        df.drop(x, inplace=True) 

df 
```

```python
for x in df.index:
    if df.loc[x, 'Phone_Number'] == '':
        df.drop(x, inplace=True)

df 
```

```python
df.reset_index(drop=True, inplace=True)  # Resetting the index after dropping rows

df 
```
